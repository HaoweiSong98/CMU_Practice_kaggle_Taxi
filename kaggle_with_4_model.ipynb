```python
train_data.day_of_week = train_data.day_of_week.map(lambda x:change_dow(x))
test_data.day_of_week = test_data.day_of_week.map(lambda x:change_dow(x))
```

```python
X = train_data.iloc[:,2:10]
y = train_data.iloc[:,1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=45)
```

## KNN

```python
knn = KNeighborsRegressor(n_jobs=-1, n_neighbors=5, weights='distance')
knn.fit(X_train, y_train)
```

```python
y_pred_knn = knn.predict(X_test)
y_pred_knn = np.where(y_pred_knn<0, 0, y_pred_knn)
```

```python
# RMSE score on training data for KNN
rmse = mean_squared_error(y_test, y_pred_knn, squared=False)
print("RMSE score on training data for KNN: {:.4f}".format(rmse))
```

## Gradient Boosting

```python
# Parameters for Gradient Boosting Regressor
gb = GradientBoostingRegressor(learning_rate=0.2, 
                               random_state=3, 
                               n_estimators=180, 
                               subsample=0.78, 
                               max_depth=5)

# Train Gradient Boosting Regressor
gb.fit(X_train, y_train)
```

```python
# Predict using X_test in training dataset
y_pred_gb = gb.predict(X_test)
y_pred_gb = np.where(y_pred_gb<0, 0, y_pred_gb)
```

```python
# RMSE score on training data for Gradient Boosting
rmse = mean_squared_error(y_test, y_pred_gb, squared=False)
print("RMSE score on training data for Gradient Boosting: {:.4f}".format(rmse))
```
